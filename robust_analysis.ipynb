```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        e=line.strip()
        col=line.split()
        data.append(col)	
    return data
```

```python
def pipiline(lista):
    data_float=[]
    for i in lista:
        data_float.append([float(i[2]), float(i[3]), float(i[4]), float(i[5]), float(i[6]), float(i[7])])
    mean_data = []
    std_data = []
    for i in data_float:
        mean_data.append(np.mean(i))
        std_data.append(np.std(i))
    return mean_data, std_data
```
